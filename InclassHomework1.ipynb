```r
### 1)
# Read in the data
D <- read.table("data/djao2.csv", header=T)
B1 <- ts(D$AO, start=c(1,1), freq=1)
B2 <- diff(log(B1))*100
```

```r
# Plot the data
plot(B2)
```

```r
### To be weakly stationary it must satisfy the following conditions:
### - E(Xt) does not depend on t
### - V(Xt) does not depend on t.
### - cor(Xt, Xt+h) does not depend on t.
###
### B2 does appear to be wekly stationary.  The mean appears
### to stay around 0, the variance seems to be realitively consistent
### and not vary with t.  
###
#####################################################################
```

```r
### 2)
### 
### Plat ACF of B2
acf(B2)
```

```r
### B2 does look like white noice.  The correlation value for all 
### of the lags are within the blue dotted lines, which represent
### the 95% confidence interval of 0 ACF.
###
##################################################################

### 3)

### The blue dashed lines in the ACF plot represent the 95% confidence
### interval under no autocorrelation.  
n = length(B2)
CU.u = 1.96/sqrt(n)
CU.l = -1.96/sqrt(n)
c(CU.l, CU.u)
```

```r
### Confidence interval is 0.12396, -0.12396
### Check the numbers
acf(B2)
abline(h=c(CU.u, CU.l), col="red")
```

```r
### The blue lines represent the 95% confidence interval of
### white noise (the ACF is 0).  The values are 0.117 and -0.117.
###
#################################################################

### 4)
mean(B2)
```

```r
sd(B2)
```

```r
### sample mean is 0.02702
### sample standard deviation is 0.8806
##############################################

### 5)
```

```r
CI.u = mean(B2) + 1.96 * sd(B2)*sqrt(1/n)
CI.l = mean(B2) - 1.96 * sd(B2)*sqrt(1/n)
c(CI.u, mean(B2), CI.l)
```

```r
### H0: True mean is 0
### Ha: True mean is not 0
### Confidence level 95%
### Confidence interval is -0.0821 to 0.1362.  Zero is within
### the confidence interval, so we fail to reject the null
## hypothesis that the true mean is 0.

################################################################
```
