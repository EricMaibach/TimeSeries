Theoretical ACF and ACVF of ARMA

```r
phis <- c(0.4, 0.2)
thetas <- c(0.6, 0.2)
```

```r
Th.acf <- ARMAacf(ar=phis, ma=thetas, lag.max=20)
Th.pacf <- ARMAacf(ar=phis, ma=thetas, lag.max=20, pacf=T)
```

```r
mu <- 5
x <- arima.sim(n = 250, list(ar = phis, ma = thetas)) + mu
```

```r
plot(x, type="o")
```

```r
acf(x)
lines(0:20, Th.acf, type='p', col="red")
```

```r
pacf(x)
lines(1:20, Th.pacf, type="p", col="red")
```
