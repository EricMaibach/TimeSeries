# Theoretical ACF and ACVF of ARMA

```r
phis <- c(0.4, 0.2)
thetas <- c(0.6, 0.2)
```

```r
Th.acf <- ARMAacf(ar=phis, ma=thetas, lag.max = 20)
```

```r
Th.pacf <- ARMAacf(ar=phis, ma=thetas, lag.max = 20, pacf=T)
```

```r
mu <- 5
x <- arima.sim(n = 250, list(ar = phis, ma = thetas)) + mu
```

```r
plot(x, type="o")
```

```r
acf(x)
lines(0:20, Th.acf, type="p", col="red")
```

```r
pacf(x)
lines(1:20, Th.pacf, type="p", col="red")
```

# Example: Annual Copper Price

```r
library(forecast)
```

```r
D <- read.csv("https://nmimoto.github.io/datasets/copper.csv")
D1 <- ts(D[,2], start=c(1800,1), freq=1)
plot(D1, type='o')
```

```r
acf(D1)
```

```r
pacf(D1)
```

```r
Fit01 <- auto.arima(D1, d=0, approximation = FALSE)
Fit01
```

```r
Fit02 <- auto.arima(D1, trace = TRUE, d=0, approximation = FALSE)
```

```r
Fit02
```

```r
Fit03 <- auto.arima(D1, max.order=5, stepwise = FALSE, trace=TRUE, d=0, approximation = FALSE)
```

```r
Fit03
```

```r
Arima(D1, order=c(2,0,3))
```
